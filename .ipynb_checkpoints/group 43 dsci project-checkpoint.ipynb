## **Introduction**

Understanding what drives player engagement is an important question in mobile gaming research, especially when measuring voluntary actions such as newsletter subscription. When a player signs up for the newsletter of a game they are playing, it often indicates a deeper interest in a game or its community. Rashed et al. (2025) highlights that player engagement is multidimensional, involving cognitive, emotional, and behavioral components. Because engagement cannot be observed directly, it is often measured through behavioral indicators such as session length, frequency of play, and retention rates, which provide insights into how invested players are in a game environment.

In this project, we analyze data from a Minecraft research server made by a research group at UBC. The server allows the research team to explore how players behave in digital games, giving researchers insight into how players explore, interact, and invest time in a virtual environment. By analyzing both demographic information and gameplay behavior, we aim to determine whether individual characteristics can help predict which players are more likely to subscribe to the game’s newsletter.

The broad question utilized in our project is:
"What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types? “

To refine this question for analysis, we focus on the specific question:
 **“Does the average session length and age predict whether a player subscribes to the newsletter?”**

## **Methods & Results**

In order to solve our specific question, we will use a KNN-Classification model, in order to create a predictive model regarding somone's likelyhood to subscribe based on their Age and Session play time. The predictive model along with our data will determine the accuracy of our prediction and whether using the two predictors to determine subscription probability would have a high accuracy.

```r
#loads data using read_csv function
players_origin <- read_csv("https://raw.githubusercontent.com/mcheng250/DSCI_Project_final_report/refs/heads/main/data/players.csv")
sessions_origin <- read_csv("https://raw.githubusercontent.com/mcheng250/DSCI_Project_final_report/refs/heads/main/data/sessions.csv")
```

**Clean the sessions.csv:**

In order to clear our data, we would need to mutate start_time and end_time to get session_length by using dmy_hm to turn start_time and 
end_time into proper format, use difftime to calculate session length and wrangle the session_length so we have avg_session_length. 

We would also clean other columns so all we left is hashedEmail and avg_session_length while rounding the avg_session_length 
decimal place so the data looks more clean

```r
#Clean the data through mutate, difftime, and summarize
sessions_tidy <- sessions_origin |>
                    mutate(start_time = dmy_hm(start_time),
                           end_time = dmy_hm(end_time),
                           session_length = as.numeric(difftime(end_time, start_time, units = "mins"))) |>
                    group_by(hashedEmail) |>
                    summarize(avg_session_length = round(mean(session_length, na.rm = TRUE),2))
head(sessions_tidy)
```

```r
# Take out age,subscribe and hashedEmail from the player dataset to clean up players.csv
players_tidy <- players_origin |>
                    select(Age,subscribe,hashedEmail)  
head(players_tidy)
```

```r
# Merge two datasets to make the final clean data
tidy_data <- players_tidy |>
                left_join(sessions_tidy, by = "hashedEmail") |>
                drop_na(avg_session_length)
head(tidy_data)
```

**Summary of dataset:**

```r
# The summary of mean value, min value and max value of avg_session_length
tidy_data |> summarize(mean_avg_session_length = round(mean(avg_session_length, na.rm = TRUE),2),
                       min_avg_session_length = round(min(avg_session_length, na.rm = TRUE),2),
                       max_avg_session_length = round(max(avg_session_length, na.rm = TRUE),2))
# The summary of mean value, min value and max value of Age
tidy_data |> summarize(mean_age = round(mean(Age, na.rm = TRUE),2),
                            min_age = round(min(Age, na.rm = TRUE),2),
                            max_age = round(max(Age, na.rm = TRUE),2))
# The summary of players whose subscribe is FALSE
tidy_data |> filter(subscribe == FALSE) |> 
                summarize(subscribe_FALSE = n())
# The summary of players whose subscribe is TRUE
tidy_data |> filter(subscribe == TRUE) |> 
                summarize(subscribe_TRUE = n())
# table of mean value summary
tidy_data |> summarize(mean_avg_session_length = round(mean(avg_session_length, na.rm = TRUE),2),
                                        mean_age = round(mean(Age, na.rm = TRUE),2)) |>
                              pivot_longer(cols = everything(),
                                           names_to = "variable", 
                                           values_to = "mean_value")
```

**KNN Classification Model & Determining Accuracy**

```r
# Convert subscribe data to factor and remove any NA values within the dataset using drop_na
tidy_data <- tidy_data |>
  mutate(subscribe = factor(subscribe, levels = c(TRUE, FALSE), 
                          labels = c("Yes", "No"))) |>   
drop_na(Age, avg_session_length, subscribe)
```

```r
set.seed(3456) 
# Randomly take 75% of the data for the training set, and 25% for testing set
data_split <- initial_split(tidy_data, prop = 0.75, strata = subscribe)  
train_data <- training(data_split)   
test_data <- testing(data_split)
# Scaling the variable and create recipe
knn_recipe <- recipe(subscribe ~ Age + avg_session_length, data = train_data) |>
            step_scale(all_predictors()) |>
            step_center(all_predictors())
# Create a knn-model specification 
knn_spec <- nearest_neighbor(weight_func = "rectangular", 
  neighbors = tune()) |>
  set_engine("kknn") |>
  set_mode("classification")
```

```r
# combine the specification and recipe into a workflow but do not fit yet 
knn_workflow <- workflow() |>
  add_recipe(knn_recipe) |>
  add_model(knn_spec)
```

```r
# 10-fold cross-validation
train_vfolds <- vfold_cv(train_data, v = 10, strata = subscribe)
```

```r
# created a grid of K values ranging from 1 to 30 to test for a optimal K value
k_values <- tibble(neighbors = seq(1, 30))

# Tune to find best k by applying the previous workflow
knn_results <- knn_workflow |>
  tune_grid(resamples = train_vfolds, grid = k_values) |>
  collect_metrics()

# Determine the k based on highest accuracy
best_k <- knn_results |>
  filter(.metric == "accuracy") |>
  arrange(desc(mean)) |>
  slice(1)

best_k
```

```r
# Visualization of K tuning results
knn_results |>
  filter(.metric == "accuracy") |>
  ggplot(aes(x = neighbors, y = mean)) +
  geom_point() +
  geom_line() +
  geom_vline(xintercept = best_k_value, linetype = "dashed", color = "red") +
  labs(x = "Number of Neighbors (K)",
       y = "Cross-Validation Accuracy",
       title = "Model Accuracy vs. K Value") +
  theme_minimal()
```

Figure 4: A visual of the models accuracy against each K-value, with a red line through the optimal k.

**Prediction models and Accuracy & Precision calculations**

```r
# pull only the neighbors value of the best K 
best_k_value <- best_k |> pull(neighbors)
```

```r
#created the final KNN model specification using the best K value. 
final_knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = best_k_value) |>
  set_engine("kknn") |>
  set_mode("classification")
```

```r
#establish the final workflow, with the knn_recipe and the final model specification that has the optimal K value
final_workflow <- workflow() |>
  add_recipe(knn_recipe) |>
  add_model(final_knn_spec)
```

```r
#fit the workflow with training data. 
final_fit <- final_workflow |>
  fit(data = train_data)
```

```r
#Using the final fitted workflow, predict the subscription status of test data
test_predictions <- final_fit |>
  predict(test_data) |>
  bind_cols(test_data)

head(test_predictions)
```

```r
# Generate confusion matrix comparing actual vs predicted subscription values
test_predictions |>
  conf_mat(truth = subscribe, estimate = .pred_class)

#Calculate performance metrics and filter to extract the accuracy metric
test_accuracy <- test_predictions |>
  metrics(truth = subscribe, estimate = .pred_class) |>
  filter(.metric == "accuracy")

#Calculate the proportion of predicted subscribers that were actually subscribers
test_precision <- test_predictions |>
  precision(truth = subscribe, estimate = .pred_class)

test_accuracy
test_precision
```

```r
cat("Best K:", best_k_value, "\n")
cat("Cross-validation accuracy:", round(best_k$mean, 3), "\n")
cat("Test set accuracy:", round(test_accuracy$.estimate, 3), "\n")
```

By analyzing the data, we are able to determine the test set accuracy and cross-validation accuracy which demonstrates the strength in predicting Subscription status based upon the Age and Average session length of player. 

```r
#Create a scatter plot of test data's predicted subscriptions against the actual subscription
prediction_plot <- test_predictions |>
ggplot(aes(x = Age, y = avg_session_length)) +
  geom_point(aes(color = .pred_class, shape = subscribe), size = 3, alpha = 0.7) +
  labs(x = "Age",
       y = "Average Session Length (minutes)",
       color = "Predicted Subscription",
       shape = "Actual Subscription",
       title = "KNN Model Predictions vs Actual Subscription Status (of test data) ",
       subtitle = paste("K =", best_k_value)) +
  scale_color_manual(values = c("Yes" = "blue", "No" = "red")) +
  theme_minimal() +
  theme(legend.position = "right")

prediction_plot
```

## **Discussion**

**Summary of Findings** 

Our project explored whether player age and average session length could predict newsletter subscription on a Minecraft research server. Using a cleaned dataset linking demographic data to gameplay behavior, we trained and tuned a KNN classification model. 
Cross-validation identified K = 7 as the optimal number of neighbors, giving a cross-validation accuracy of ~0.754 and a test accuracy of 0.71. This suggests that the model generalizes reasonably well to unseen data. Visualization of the predictions showed a clear pattern: players with longer session lengths were much more likely to be predicted as subscribers, while shorter or irregular session times corresponded more strongly with non-subscribers. Most incorrect predictions occurred in regions where the two classes overlapped, meaning the model struggled when the behavioral differences between groups were subtle.


**Was This Expected?** 

Our findings matched what we expected. In the introduction, we explained that signing up for a newsletter is a deeper form of engagement. Earlier research shows that behaviour within a session can reveal how invested a player is in a game. For this reason, it made sense that average session length would be the strongest predictor, and our results supported this. Age had a small effect, which was unexpected, but it did not separate the two groups in a clear way. This suggests that players of many ages act in a similar way when they choose to subscribe. It also shows that behaviour gives stronger clues about engagement than demographic information.

**Impact of the Findings**

These results have several important implications for understanding and supporting player engagement. They show the value of prediction models that focus on behaviour. If actions related to engagement, such as subscribing, are best explained by how players behave, then analytics should place more weight on behavioural measures when designing targeted features. Developers and researchers can use insights from this model to identify players who appear close to subscribing, such as players with moderately long sessions who may respond to simple prompts, personalized messages, or small incentives. The results also show that splitting players into demographic groups, such as older and younger players, is less useful for predicting or influencing newsletter sign-ups. Focusing on players who already spend more time in the game provides a clearer and more practical way to understand engagement and to design strategies that support it.

**Future Questions**

Our analysis suggests several directions for future work. Adding more behavioural features, such as total playtime, login frequency, progress through game tasks, or the times of day players are active, could improve predictions and make the model more accurate. It would also be useful to test other classification methods like logistic regression to see whether they capture relationships that KNN may miss. Another important direction is to examine changes in behaviour over time, such as whether players who show longer sessions as they continue to play are more likely to subscribe. Expanding the study to include outcomes like retention or churn could also give a more complete view of engagement on research servers and help shape strategies for supporting healthy gaming communities.
